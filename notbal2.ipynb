```python
from astropy.io import fits
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fits_image_filename = fits.open('DR12_noBAL.fits')
```

```python
noBAL = fits_image_filename
noBAL.info()
```

```python
from astropy.table import Table
NOTBAL = Table.read(fits_image_filename, hdu=1)
print(NOTBAL.columns)
```

```python
data = noBAL[1].data
```

```python
plt.figure(num=3, figsize=(18, 3),)
plt.ylim((-10, 10))
plt.plot(data[15][0], data[15][1])
plt.show()
```

```python
print(data[100][0])
print(data[100][1])
```

```python
from scipy.interpolate import interp1d
import pylab as pl
```

```python
import pylab as pl
```

```python
pl.figure(num=3, figsize=(18, 3),)
wave=data[1][0]
flux=data[1][1]
f1=interp1d(wave,flux,kind='linear')
waveint=np.linspace(data[1][0][0],data[1][0][-1],2048)
pl.plot(waveint,f1(waveint))
pl.legend(loc="best")
pl.show()
```

```python
pl.figure(num=3, figsize=(18, 5),)
pl.ylim((-10,10))
wave=data[1][0]
flux=data[1][1]
f1=interp1d(wave,flux,kind='linear')
waveint=np.linspace(data[1][0][0],data[1][0][-1],2048)
pl.plot(waveint,f1(waveint))
pl.legend(loc="best")
pl.show()
```
